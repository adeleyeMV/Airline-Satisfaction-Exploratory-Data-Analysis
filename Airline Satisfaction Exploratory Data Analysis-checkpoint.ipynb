**Column have outliers we try to fixed it** 

```python
# create function that remove outliers of numiric column

def remove_outliers(column_name):
    Q1 = data[column_name].quantile(0.25)

    Q3 = data[column_name].quantile(0.75)

    IQR = Q3 - Q1

    Upper_boundary = Q3 + (1.5 * IQR)

    Lower_boundary = Q1 - (1.5 * IQR)
    
    # replace number > Upper_boundary and number < Lower_boundary with nan value
    data.loc[(data[column_name] > Upper_boundary) | (data[column_name] < Lower_boundary), column_name] = np.nan
    
    # replace nan value with mean
    data[column_name].fillna(data[column_name].mean(), inplace = True)
```

```python
# remove outlier 

remove_outliers("Departure Delay") # call function i create it in cell 25
```

```python
# see distribution after cleaning

sns.boxplot(y = "Departure Delay", data = data)
plt.show()
```

## Bivariate Analysis & Visualizations

### 1) What is Age distribution With categorical columns ?

```python
# 1- create list of categorical column
cat_column = data.select_dtypes(include = object).columns.tolist()
```

```python
# create function that visualized catigorical column with age column

def create_kdeplot(x_axis, columns):
    # 1- set figure size
    plt.figure(figsize=(15, 15))

    # 2- loop over categorical column list to plot columns
    for index, col in enumerate(columns):
        plt.subplot((len(columns) + 1) // 2, 2, index + 1) # create sub-plot
        sns.kdeplot(x = x_axis, hue = col, data = data, fill = True)

        plt.title(col) # set title to each plot
        plt.xlabel("") # replace x label with empty string
        plt.ylabel("") # replace y label with empty string
        plt.yticks([]) # Remove y-axis label
 
    # 3- set layout between two plots
    plt.tight_layout(pad = 2)

    plt.show()
```

```python
# call function i defined in cell 46
create_kdeplot("Age", cat_column) # cat_column list from cell 45
```

- Passengers of older ages:

### 4) What is Relation between  average  services Rating and categorical columns?

```python
# create function that Grouped specific column by services

def plot_services(column_name):
    services = pd.DataFrame(data.groupby(column_name)[services_columns].mean().round(1))
    services = services.T # transpose data frame to Vertical shape
    
    # Creat plot 
    plt.figure(figsize = (15,4)) # set size to plot
    sns.lineplot(data = services, markers = True)
    
    # set x-axis lable new position
    plt.xticks(rotation = 45)
    
    # set title
    plt.title(f'Service Rating With {column_name}')
    
    # show figure
    plt.show()
```

```python
# Visualize categorical columns with Services rating

for column in cat_column: # cat_column list i defined it in cell 45
    plot_services(column) # call function i defined it in cell 50
```

```python
# see correlation matrix between numerical columns

plt.figure(figsize = (10,6))
sns.heatmap(data.select_dtypes(exclude = object).corr(), annot = True, fmt = ".2f", linewidths = 0.2)
plt.show()
```

**There's a lot of Strong Positive Relationships between Services Ratings, showen**
